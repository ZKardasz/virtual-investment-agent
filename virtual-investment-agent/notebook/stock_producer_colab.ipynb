```python

# Importy
import yfinance as yf
import json
from datetime import datetime
import time
import pandas as pd

```

```python
# Ustawienia
TICKERS = ["AAPL", "MSFT", "TSLA"]
FETCH_INTERVAL_SECONDS = 60  # możesz zmienić na 10 na potrzeby testów
OUTPUT_FILE = "stock_data.json"

```

```python
# Funkcja do pobierania ceny
def fetch_price(ticker):
    data = yf.download(tickers=ticker, period="1d", interval="1m")
    if not data.empty:
        return float(data['Close'].iloc[-1].item())  # <- to jest w 100% poprawne
    return None
```

```python
# Główna pętla (dla testów: 3 iteracje)
all_data = []

for _ in range(3):  # wykonaj 3 razy (dla testów; normalnie byłaby nieskończona pętla)
    for ticker in TICKERS:
        price = fetch_price(ticker)
        if price is not None:
            record = {
                "timestamp": datetime.utcnow().isoformat(),
                "stock": ticker,
                "price": round(price, 2)
            }
            all_data.append(record)
            print(record)

            # Zapis do pliku JSON (dodaj linię)
            with open(OUTPUT_FILE, "a") as f:
                f.write(json.dumps(record) + "\n")
    time.sleep(FETCH_INTERVAL_SECONDS)


```

```python
# Konwersja do DataFrame i podgląd danych
df = pd.DataFrame(all_data)
df
```
